```python
import boto3
import json
import random
import time
from datetime import datetime
```

```python
# Initialize Kinesis client with correct region
kinesis = boto3.client('kinesis', region_name='us-east-1')

# Define stream name
stream_name = 'Stockdatastream'
```

```python
# Simulate stock data
def generate_stock_data():
    stock_data = {
        'symbol': 'AAPL',  # Example stock symbol
        'price': round(random.uniform(120.0, 150.0), 2),  # Random price
        'timestamp': datetime.utcnow().isoformat()  # Current UTC timestamp
    }
    return stock_data
```

```python
# Send data to Kinesis
def send_to_kinesis(data):
    payload = json.dumps(data)
    response = kinesis.put_record(
        StreamName=stream_name,
        Data=payload,
        PartitionKey='partitionKey'  # Fixed partition key for testing
    )
    print(f"Record sent: {response['SequenceNumber']}")
```

```python
record_limit = None  # Set to None for continuous streaming or specify a limit
count = 0

while record_limit is None or count < record_limit:
    # Check for stop signal
    if os.path.exists('stop_signal.txt'):
        print("Stop signal received. Stopping producer...")
        break

    # Generate and send stock data
    stock_data = generate_stock_data()
    send_to_kinesis(stock_data)
    time.sleep(2)  # Adjust delay as needed
    count += 1

print("Producer script stopped.")
```
